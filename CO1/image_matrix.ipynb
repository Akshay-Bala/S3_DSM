```python
import numpy as z
from PIL import Image
from scipy import ndimage
from matplotlib import pyplot as plt


#view image

img=Image.open("zz.jpg")
plt.imshow(img)
```

```python
#rotate image

a=z.array(img)
b=ndimage.rotate(a,180)
plt.imshow(b)
```

```python

#blur image
blur=ndimage.gaussian_filter(img,sigma=2)
plt.imshow(blur)
plt.title("Blurred Image")
```

```python

#zoom image
d=ndimage.zoom(a,(.4,.2,1))
plt.imshow(d)
plt.title("zoom image")
```
