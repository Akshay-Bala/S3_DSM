```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,10)
y=np.arange(1,10)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y,color="blue")
plt.show()
```

```python
#scatter plot
x=[1,2,3,4,5]
y=[10,12,5,8,6]
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')
plt.show()
```

```python

#histogram
data=np.random.randn(1000)
plt.hist(data,bins=20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```
